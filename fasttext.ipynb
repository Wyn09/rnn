无监督

```python
with open("../data/《过分野》作者：鸟松米.txt") as f:
    context = f.read()
splited = jieba.lcut(context)
with open("../data/xiaoshuo_splited.txt", "w", encoding="utf-8") as f:
    f.write(" ".join(splited))
```

```python
model = fasttext.train_unsupervised("../data/xiaoshuo_splited.txt", model="cbow", dim=128) # 语料不是很多时用cbow，否则用skipgram
```

```python
print(model.words)
```

```python
model.get_nearest_neighbors("姑娘")
```

词向量可视化

```python
from torch.utils.tensorboard import SummaryWriter
```

```python
writer = SummaryWriter("../source/logs")
```

```python
# 绘制词向量可视化图形

# 词汇表
meta = model.words[:200]

embeddings = []

for word in meta:
    embeddings.append(torch.from_numpy(model.get_word_vector(word)).reshape(1,-1))

word_emb = torch.cat(embeddings, dim=0)
print(word_emb.shape)
```

```python
writer.add_embedding(word_emb, metadata=meta)
```
